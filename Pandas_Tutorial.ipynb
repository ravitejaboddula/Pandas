```python
import pandas as pd

# 1. From a Python list (automatically generates numeric index 0, 1, 2)
list_series = pd.Series([10, 20, 30])
print(list_series)
# 2. From a list with a custom labeled index
custom_index_series = pd.Series([10, 20, 30], index=["Row A", "Row B", "Row C"], dtype='int8')
print(custom_index_series)

# 3. From a Python dictionary (keys automatically become the index labels)
dict_series = pd.Series({"Apple": 5, "Banana": 3, "Cherry": 10})
print(dict_series)
```

Sort values

```python
#sorting the units sold column in descending
coffee.sort_values(["Units Sold"], ascending = False)
```

```python
#sorting the units sold in decending and coffee type ascendign using binary false and true
coffee.sort_values(["Units Sold","Coffee Type"], ascending=[0,1])
```
